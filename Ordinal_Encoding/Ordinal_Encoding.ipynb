```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/shop data.csv')
df
```

Ordinal Encoder

```python
#Find unique
df['income'].unique()
```

```python
#Telling the order
income_order = ['low','medium','high']
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
enco = OrdinalEncoder(categories=[income_order])
```

```python
enco.fit(df[['income']])
```

```python
final = pd.DataFrame(enco.transform(df[['income']])) # n_samples , n_values
final
```

```python
df['income'] = final
df
```
